```python
import pandas as pd
import numpy as np

loan_data = pd.read_csv('../data/loan_data.csv')
```

```python
loan_data.head()
```

```python
loan_data['Property_Area'].value_counts()
```

```python
loan_data = pd.get_dummies(loan_data, columns=['Property_Area'], prefix=['Area'])
loan_data.head()
```

```python
Y = loan_data['Loan_Status']
loan_data = loan_data.drop(columns=['Loan_Status'])
```

```python
loan_data.head()
```

```python
Y.replace(['Y', 'N'], [1, 0], inplace=True)
```

```python
Y.head()
```

```python
loan_data.replace(['Yes', 'No'], [1, 0], inplace=True)
loan_data.head()
```

```python
loan_data.replace(['Graduate', 'Not Graduate'], [1, 0], inplace=True)
loan_data.head()
```

```python
loan_data = pd.get_dummies(loan_data, columns=['Gender'])
loan_data.head()
```

```python
loan_data.drop(columns=['Loan_ID'], inplace=True)
loan_data.head()
```
